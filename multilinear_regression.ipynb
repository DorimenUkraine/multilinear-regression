```python
import numpy as np
from bokeh.plotting import figure, output_notebook, show
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
output_notebook()
```

```python
def generate_dataset(n):
    x = []
    y = []
    for i in range(n):
        x.append([1, i, i+np.random.randint(-n/2, high=n)])
        y.append(np.random.rand())
    return np.array(x), np.array(y)
```

```python
x, y = generate_dataset(200)

mpl.rcParams['legend.fontsize'] = 12

fig = plt.figure()
ax = fig.gca(projection='3d')

ax.scatter(x[:,1],x[:,2], y, label='points', s=5)
ax.legend()
ax.view_init(45, 45)

plt.show()
```

```python
def mse(coef, x, y):
    return np.mean((np.dot(x, coef) - y)**2)
```

```python
def gradients(coef, x, y):
    return np.mean(x.transpose()*(np.dot(x, coef) - y), axis=1)
```

```python
def multilinear_regression(coef, x, y, lr, epsilon=8e-2):
    prev_error = 0
    while True:
        error = mse(coef, x, y)
        print(abs(error - prev_error))
        if abs(error - prev_error) <= epsilon:
            break
        grad = gradients(coef, x, y)
        coef = np.subtract(coef, lr*grad)
    return coef
```

```python
coef = np.array([1, x[:,1].mean(), x[:,2].mean()])
c = multilinear_regression(coef, x, y, 4e-5)
print(c)
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')

ax.scatter(x[:,1],x[:,2], y, label='points', s=5, color="red")
x1, x2 = np.meshgrid(x[:,1], x[:,2])
ax.plot_surface(x1,x2, c[0] + c[1]*x[:,1] + c[2]*x[:,2], label='regression')
# ax.legend()
ax.view_init(45, 90)

plt.show()
```
